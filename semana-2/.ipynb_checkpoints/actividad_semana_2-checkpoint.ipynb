```python
import pandas as pd
```

```python
# Carga del archivo CSV y vista previa de las primeras filas
ventas = pd.read_csv("ventas (1).csv")
print("Primeras filas del archivo de ventas:")
ventas.head()
```

```python
# Carga del archivo JSON y cantidad total de clientes
clientes = pd.read_json("clientes.json")

print("Primeras filas del archivo de clientes:")
print(clientes.head())

cantidad_clientes = len(clientes)
print("Cantidad total de clientes:", cantidad_clientes)
```

```python
# Carga del archivo Excel y promedio de stock disponible
inventario = pd.read_excel("inventario.xlsx")

print("Primeras filas del archivo de inventario:")
print(inventario.head())

promedio_stock = inventario["stock"].mean()
print("Promedio de stock disponible:", promedio_stock)
```

## Conclusiones:

### Archivo CSV (ventas)
Se cargaron correctamente los datos de ventas y se pudo visualizar su estructura, lo que permite entender las variables disponibles para futuros análisis.
### Archivo JSON (clientes)
Se importaron los datos de clientes y se obtuvo la cantidad total, lo que permite conocer el volumen de clientes registrados.
### Archivo Excel (inventario)
Se cargaron los datos de inventario y se calculó el promedio de stock, lo que permite tener una referencia general del nivel de productos disponibles.
