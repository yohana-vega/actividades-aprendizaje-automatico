```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Automobile.csv")
df.head()
```

```python
plt.scatter(df["mpg"], df["weight"], c="m")
plt.xlabel("mpg")
plt.ylabel("weight")
plt.title("Relación entre mpg y weight")
plt.show()
```

```python
plt.hist(df["mpg"], bins=15, color="skyblue", edgecolor="black")
plt.xlabel("mpg")
plt.ylabel("Frecuencia")
plt.title("Distribución del consumo (mpg)")
plt.show()
```

```python
plt.boxplot([
    df[df["origin"] == "europe"]["cylinders"],
    df[df["origin"] == "japan"]["cylinders"],
    df[df["origin"] == "usa"]["cylinders"]
])

plt.xticks([1, 2, 3], ["europe", "japan", "usa"])
plt.xlabel("Origen")
plt.ylabel("Cantidad de cilindros")
plt.title("Cantidad de cilindros según el origen")
plt.grid()
plt.show()
```

```python
conteo = df.groupby(["model_year", "origin"]).size().unstack()

conteo.plot(kind="bar")

plt.xlabel("Año del modelo")
plt.ylabel("Cantidad de autos")
plt.title("Distribución de autos por año y origen")
plt.legend(title="Origen")
plt.grid()

plt.show()
```

**Análisis del gráfico**


En el gráfico de dispersión se observa una relación inversa entre el peso del vehículo y el consumo (mpg), ya que los autos más pesados tienden a consumir más combustible.

En el histograma, la mayoría de los valores de mpg se concentran entre 15 y 30, lo que indica un rendimiento medio en la mayoría de los vehículos.

En el boxplot, se observa que los autos de origen USA presentan mayor variabilidad en la cantidad de cilindros y valores más altos, mientras que Europa y Japón tienen menor dispersión.

En el gráfico de barras, se puede ver cómo varía la cantidad de autos según el año y el origen, destacándose una mayor presencia de vehículos de USA en la mayoría de los años.
