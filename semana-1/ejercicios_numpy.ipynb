```python
# Ejercicio 1: Crear un array de 10 elementos con ceros
import numpy as np

array_ceros = np.zeros(10)
print(array_ceros)
```

```python
# Ejercicio 2: Generar un array con los números del 1 al 20
array_1_20 = np.arange(1, 21)
print(array_1_20)
```

```python
# Ejercicio 3: Crear un array con los números pares del 2 al 50
array_pares = np.arange(2, 51, 2)
print(array_pares)
```

```python
# Ejercicio 4: Crear una matriz de 4x4 con valores aleatorios entre 0 y 1
matriz = np.random.rand(4, 4)
print(matriz)
```

```python
# Ejercicio 5: Suma elemento a elemento de dos arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

suma = a + b
print(suma)
```

```python
# Ejercicio 6: Multiplicar dos matrices 3x3 de enteros aleatorios entre 1 y 10
A = np.random.randint(1, 11, (3, 3))
B = np.random.randint(1, 11, (3, 3))

producto = np.dot(A, B)
print(producto)
```

```python
# Ejercicio 7: Crear un array de 15 valores aleatorios entre 1 y 100
array = np.random.randint(1, 101, 15)

print("Array:", array)
print("Máximo:", np.max(array))
print("Mínimo:", np.min(array))
print("Promedio:", np.mean(array))
```

```python
# Ejercicio 8: Generar una matriz 5x5 del 1 al 25 y extraer su diagonal
matriz = np.arange(1, 26).reshape(5, 5)

diagonal = np.diag(matriz)

print("Matriz:\n", matriz)
print("Diagonal:", diagonal)
```

```python
# Ejercicio 9: Matriz 5x5 aleatoria y suma por filas y columnas
matriz = np.random.randint(1, 21, (5, 5))

suma_filas = matriz.sum(axis=1)
suma_columnas = matriz.sum(axis=0)

print("Matriz:\n", matriz)
print("Suma por filas:", suma_filas)
print("Suma por columnas:", suma_columnas)
```
