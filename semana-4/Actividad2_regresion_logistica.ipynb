# Actividad 2 - Regresión Logística

## Introducción

En esta actividad se trabaja con un dataset que contiene información sobre el comportamiento de usuarios en un sitio web.

El objetivo es predecir qué sistema operativo utiliza el usuario (Windows, Macintosh o Linux) a partir de variables como duración de la visita, páginas vistas y acciones realizadas.

Para ello se aplica un modelo de regresión logística, que permite clasificar los datos en categorías discretas.

```python
# Importamos pandas
import pandas as pd

# Cargamos el dataset
df = pd.read_csv("usuarios_win_mac_lin.csv")

# Mostramos las primeras filas
df.head()
```

```python
# Definimos variables
X = df.drop("clase", axis=1)  # variables de entrada
y = df["clase"]               # variable objetivo
```

```python
# Importamos modelo
from sklearn.linear_model import LogisticRegression

# Creamos modelo (ajustado)
modelo = LogisticRegression(max_iter=1000)

# Entrenamos
modelo.fit(X, y)
```

## Interpretación de la variable objetivo

La variable "clase" representa el sistema operativo del usuario:

0 = Windows  
1 = Macintosh  
2 = Linux  

El modelo utiliza estas categorías para clasificar a cada usuario según su comportamiento en el sitio web.

```python
# Predicción sobre los mismos datos
y_pred = modelo.predict(X)

# Mostramos resultados
print(y_pred)
```

```python
# Evaluamos el modelo
score = modelo.score(X, y)

print("Precisión del modelo:", score)
```

```python
# Importamos matriz de confusión
from sklearn.metrics import confusion_matrix

# Creamos la matriz de confusión
matriz = confusion_matrix(y, y_pred)

# Mostramos la matriz
print(matriz)
```

```python
import matplotlib.pyplot as plt

plt.imshow(matriz)
plt.title("Matriz de confusión")
plt.xlabel("Predicción")
plt.ylabel("Valor real")
plt.colorbar()
plt.show()
```

## Conclusión

Se aplicó un modelo de regresión logística para clasificar el sistema operativo de los usuarios a partir de variables como duración de la visita, páginas vistas, acciones y valor.

El modelo obtuvo una precisión aproximada del 77%, lo que indica un desempeño aceptable.

A partir de la matriz de confusión se observa que el modelo presenta algunos errores de clasificación entre las categorías, aunque logra identificar correctamente una gran parte de los casos.

Esto demuestra que las variables utilizadas permiten captar patrones de comportamiento, pero podrían mejorarse para aumentar la precisión del modelo.
